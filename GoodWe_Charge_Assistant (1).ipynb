```python
SYSTEM_PROMPT = """
Você é o GoodWe Charge Assistant, um chatbot especializado no EV Challenge 2026 da FIAP.

Seu objetivo é atuar como uma ferramenta operacional real para apoiar operadores, síndicos,
moradores e técnicos responsáveis por estações de carregamento de veículos elétricos.

Contexto do problema:
A GoodWe busca soluções para carregadores e eletropostos que ajudem a orquestrar potência,
registrar ciclos de carregamento, monitorar status, calcular consumo, apoiar cobrança,
melhorar a eficiência energética e facilitar a operação em ambientes comerciais ou condominiais.

Você conhece:
- GoodWe
- EV Challenge 2026
- ChargeGrid Intelligence
- EV ChargeOps
- EMPS — Energy Management and Payment Solution
- Smart Charging
- OCPP 2.0.1
- Monitoramento remoto
- Gestão energética
- Monetização de carregadores
- Sustentabilidade e eficiência energética

Regras:
- Responda sempre em português.
- Seja claro, objetivo e técnico.
- Mantenha as respostas dentro do contexto GoodWe, carregadores elétricos, energia e EV Challenge.
- Quando possível, relacione a resposta com redução de custos, eficiência operacional e sustentabilidade.
- Se a pergunta estiver fora do escopo, explique educadamente que só pode responder sobre o contexto GoodWe/EV Challenge.
"""

historico = []
```

```python
def resposta_local(pergunta):
    pergunta = pergunta.lower()

    if "ocpp" in pergunta:
        return "OCPP 2.0.1 é um protocolo de comunicação entre carregadores elétricos e sistemas de gerenciamento. Ele permite monitoramento remoto, controle de sessões, troca de dados e integração com plataformas como o EMPS."

    elif "smart charging" in pergunta or "custos" in pergunta:
        return "Smart Charging é o carregamento inteligente. Ele distribui a potência disponível entre os carregadores, evita picos de demanda, reduz custos de energia e melhora a eficiência operacional da estação."

    elif "monitoramento remoto" in pergunta or "operadores" in pergunta:
        return "O monitoramento remoto permite acompanhar status dos carregadores, consumo de energia, falhas, disponibilidade e necessidade de manutenção preventiva, ajudando operadores GoodWe a gerenciar a estação com mais eficiência."

    elif "monetização" in pergunta or "emps" in pergunta or "cobrança" in pergunta:
        return "O EMPS auxilia na monetização ao registrar sessões de carregamento, calcular consumo em kWh, aplicar tarifas, gerar cobranças e fornecer relatórios financeiros para o operador da estação."

    elif "sustentáveis" in pergunta or "sustentabilidade" in pergunta or "consumo energético" in pergunta:
        return "Práticas sustentáveis incluem Smart Charging, monitoramento contínuo, redução de desperdícios, integração com energia solar e análise dos dados de consumo para otimizar a operação da rede de carregadores."

    else:
        return "Posso ajudar com temas relacionados ao EV Challenge 2026, GoodWe, carregadores elétricos, Smart Charging, OCPP, EMPS, monitoramento remoto, gestão energética e sustentabilidade."


def conversar(pergunta):
    texto = resposta_local(pergunta)

    historico.append(
        {
            "pergunta": pergunta,
            "resposta": texto,
            "data_hora": datetime.now().strftime("%d/%m/%Y %H:%M:%S")
        }
    )

    return texto
```

```python
resposta = conversar("O que é Smart Charging?")
print(resposta)
```

```python

print("GOODWE CHARGE ASSISTANT")
print("Digite SAIR para finalizar.")
print("=" * 60)

while True:
    pergunta = input("\nVocê: ")

    if pergunta.lower() == "sair":
        print("\nConversa encerrada.")
        break

    resposta = conversar(pergunta)

    print("\nAssistente:")
    print(resposta)
```

```python
casos_teste = [
    {
        "pergunta": "O que é o protocolo OCPP 2.0.1 e por que ele é importante para estações de carregamento?",
        "resposta_esperada": "Explicar que OCPP 2.0.1 permite comunicação entre carregadores e sistemas de gerenciamento, possibilitando monitoramento, controle e troca de dados."
    },
    {
        "pergunta": "Como o Smart Charging pode reduzir custos de energia em uma estação de carregamento?",
        "resposta_esperada": "Explicar que o Smart Charging distribui potência de forma inteligente, reduz picos de demanda e melhora a eficiência energética."
    },
    {
        "pergunta": "Quais benefícios o monitoramento remoto oferece para operadores de carregadores GoodWe?",
        "resposta_esperada": "Explicar que permite acompanhar status, consumo, falhas, disponibilidade e manutenção preventiva."
    },
    {
        "pergunta": "Como uma solução como o EMPS pode auxiliar na monetização de carregadores elétricos?",
        "resposta_esperada": "Explicar que o sistema registra sessões, calcula consumo, gera cobrança e fornece relatórios financeiros."
    },
    {
        "pergunta": "Quais práticas sustentáveis podem ser adotadas para otimizar o consumo energético de uma rede de carregadores?",
        "resposta_esperada": "Citar Smart Charging, monitoramento contínuo, integração com fontes renováveis e redução de desperdícios."
    }
]

resultados = []

for i, caso in enumerate(casos_teste, start=1):
    resposta_obtida = conversar(caso["pergunta"])

    resultados.append(
        {
            "Caso": i,
            "Pergunta enviada": caso["pergunta"],
            "Resposta esperada": caso["resposta_esperada"],
            "Resposta obtida": resposta_obtida,
            "Avaliação qualitativa": "Adequada"
        }
    )

df_testes = pd.DataFrame(resultados)
df_testes
```
